# Projet Banque de chants

### Il s'agit d'une banque de chants num√©rique o√π sont enregistr√©es, stock√©es les    partittions musicales selon leur usage liturgique dans l'√©glise catholique. Un utilisateur a donc la possibilt√© d'interagir avec elle, en ajoutant, en supprimer, en modifiant les partitions en fonction de leurs cat√©gories.

```python
class Partition:
    """Classe <Partition> qui cr√©e les instances de partitions des chants avec leurs titres, 
        leurs tonalit√©s, leur usage liturgique..."""
    def __init__(self, titre, tonalit√©, usage_liturgique):
        self.titre = titre
        self.tonalit√© = tonalit√©
        self.usage_liturgique = usage_liturgique
        
    def __str__(self):
        string = '<Titre>: {}\n<Tonalit√©>: {}\n<Usage Litrugique>: {}'.format(self.titre,
                                                                        self.tonalit√©, 
                                                                        self.usage_liturgique)
        return string
    
def laquelle(liste_partitions, titre_partition, p_usage_liturgique):
    """Fonction  qui recherche une partition dans une liste de partitions en fonction 
        de son titre et son usage liturgique. Elle retourne la partition  si le titre 
        et l'usage liturgique entr√© corrrespondent bien ---retourne Rien au cas contraire"""
    for partition in liste_partitions:
        if partition.titre == titre_partition and partition.usage_liturgique == p_usage_liturgique:
            return partition
        return None
    
def play():
    partitions = []
    invite_de_base = """
===========================================================
1.Quitter
2.Ajouter    <titre_partition> <tonalit√©_partition> 
             <usage_liturgique_partition>        
3.Supprimer  <titre_partition> <usage_liturgique_partition>
4.Afficher   <titre_partition> <usage_liturgique_partition>
5.Liste      <usage_liturgique_partition>

üëâChoix:
        """
    feedback = ''
    
    while True:
        action = input(feedback+'\n'+invite_de_base).lower()
        mots = action.split()
        feedback = ''

        if len(mots)>0:
            commande = mots[0]
        else:
            commande = None

        if commande == 'quitter':
            confirmation = input("voullez-vous vraiment quitter? oui/Non: " ).lower() # ici l'utlisateur
                                                          # a la possiblit√© d'annuler
                                                                    # le choix
            if confirmation == 'oui':
                print('Sortie...')
                return True
            pass
        elif commande == 'ajouter' and len(mots)>3:
            if laquelle(partitions, mots[1], mots[3]):
                feedback += 'Vous avez d√©j√† une partition portant ce titre.'
            else:
                partitions.append(Partition(mots[1], mots[2], mots[3]))
                feedback += 'Super! La partition a √©t√© ajout√©e avec succ√®s.'
        elif commande == 'supprimer' and len(mots)>2:
            for partition in partitions:
                if laquelle(partitions, mots[1], mots[3]):
                    partitions.remove(partition)
                    feedback += 'Succ√®s ! Partition supprim√©e.'
                else:
                    feedback += 'Partition non reconnu.'
        elif commande == 'afficher' and len(mots)>2:
            partition = laquelle(partitions, mots[1], mots[2])
            if partition:
                feedback += partition.__str__()
                print(partition)
            else:
                feedback += 'Partition non reconnu.'
        elif commande == 'liste' and len(mots)>1:
            for partition in partitions:
                if partition.usage_liturgique == mots[1]:
                    print('{} ({})'.format(partition.titre, partition.usage_liturgique))
        else:
            feedback += "Donn√©e(s) incorrecte. Veuillez r√©essayer s'il vous pla√Æt"
                
play()
```
